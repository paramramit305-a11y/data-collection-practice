```python
import requests
```

```python
URL = "https://api.crossref.org/works"


res = requests.get(URL)
```

```python
json_data = res.json()
```

```python
import pandas as pd

df = pd.json_normalize(json_data["message"]["items"])
df = df[["publisher", "type", "page", "publisher-location"]]
```
