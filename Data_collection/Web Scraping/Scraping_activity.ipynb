```python
import requests
from bs4 import BeautifulSoup
```

```python
# Fetch data & store it

page_count = 1

while True:
    URL = f"http://quotes.toscrape.com/page/{page_count}/"
    res = requests.get(URL)

    Soup = BeautifulSoup(res.text, "lxml")
    quotes = Soup.select("div.quote")

    if not quotes:
        print("no valid pages anymore..")
        break

    with open(f"scraped_data/quote{page_count}.html", "w", encoding="utf-8") as f:
        f.write(res.text)
        print(f"downloaded data from page{page_count}")

    page_count = page_count + 1
    
```

```python
life_quotes = []

for page in range(1, page_count):

    with open(f"scraped_data/quote{page}.html", "r", encoding="utf-8") as f:
        html_content = f.read()

    Soup = BeautifulSoup(html_content, "lxml")
    quotes = Soup.select("div.quote")

    for q in quotes:
        all_tags = []

        for tag in q.select(".tags .tag"):
            all_tags.append(tag.get_text())
            
        if "life" in all_tags:
            text = q.select_one("span.text").get_text()
            author = q.select_one("small.author").get_text()
            life_quotes.append([text, author])

    print(f"extracted life quotes from page {page}")

print("Total life quotes:", len(life_quotes))
```

```python
import pandas as pd

df = pd.DataFrame(life_quotes, columns=["text", "author"])
df
```
